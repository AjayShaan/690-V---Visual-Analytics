The following is a visualization of total expenditure on health made per capita in 260 countries in a span of 15 years starting from 1995. This dataset was downloaded from gapminder.org.

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.plotting import figure, output_file, show
import pandas as pd
from bokeh.models import ColumnDataSource

output_notebook()

health_exp = pd.read_csv("Health Expenditure.csv", encoding = "ISO-8859-1")
health_exp = health_exp.fillna(0)
health_exp
```

The dataset has been imported into the pandas dataframe health_exp, and all NaN values have been replaced by zeroes.

```python

countries = health_exp['Per capita total expenditure on health'].tolist()

years = health_exp.columns.values.tolist()[1:]

data = {}
data['countries'] = countries
for year in years:
    data[year] = health_exp[year].tolist()


p = figure(y_range=countries, plot_height=1000, plot_width = 1000, x_range=(0,10000), title="Per capita total expenditure on health")

p.hbar_stack(years, y='countries', height=0.9, color = ['red','blue','green','yellow','black','brown','olive','beige','orange','pink','khaki','grey','navy','teal','magenta','turquoise'], source=ColumnDataSource(data), legend=["%s " % x for x in years])

p.y_range.range_padding = 0.1
p.ygrid.grid_line_color = None
p.legend.location = "center_left"
p.legend.click_policy="hide"


show(p)
```

The visualization above is a cluster of stacked horizontal bar glyphs and as an interactive feature, individual glyphs can be hidden or be made to appear at the click of button by interaction with the elements in the legend.

From this visualization, it's very intuitive to infer cumulative expenditure on health per capita starting in 1995, of a country.
For example, Denmark's cumulative expenditure per capita in the year 2001 is a little over 15500.

In a similar fashion, to infer a specific year's datum associated with a country, just gauge the length of the appropriate colored horizontal bar.
For example, Canada's expenditure on health per capita in 2001 is approximately 2700.

To drill down on a specific year or to hide a certain year's data, please hide/unhide them by interacting with the legend.
